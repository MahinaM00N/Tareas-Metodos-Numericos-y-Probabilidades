Grafique en una misma figura la siguiente función junto con su serie de Taylor alrededor de $x = 0$ truncada al término número 10. ¿Hasta cuál valor de $x$ parece ser una buena aproximación?

$$
f(x) = x\sin(x)
$$

Use colores y estilos de línea diferentes para cada línea. Escoja un rango de valores de $x$ que represente lo que quiera concluir sobre esta aproximación. Incluya una leyenda para facilitar la lectura del gráfico.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1.7*np.pi,100) #para buscar un rango de x que represente una buena proximación se rebe variar este parametro, el limite superior especialmente.
y1= x*np.sin(x) #nuestro f(x)
y2=(x**2) - (1/6 *(x**4)) + (1/120 *(x**6)) - (1/5040 *(x**8)) + (1/362880 * (x**10)) #serie de taylor de xsin(x) centrada en 0 

plt.plot(x, y1, 'c--' ,label='$f(x)=xsin(x)$') #gráfico de f(x) en línea segmentada color cian 
plt.plot(x, y2,'m-', label='Serie de Taylor de f(x)')#gráfico de la serir de Taylor para f(x) en línea completa color magenta 
plt.legend() #leyenda

#extras
plt.title('Aproximación por Serie de Taylor a $f(x)= xsin(x)$') #título
plt.xlabel('$x$') # eje x
plt.ylabel('$f(x)$ y Serie de Taylor de $f(x)$ ') #eje y
plt.grid(color='b', linestyle='--', linewidth=0.2) # grilla que nos permitirá identificar valores
plt.xticks([0.5*np.pi,np.pi,1.2*np.pi,1.5*np.pi, 1.7*np.pi]) # valores de x relevantes en el gráfico

plt.show()
```

Como vemos en el gráfico ambas líneas, la correspondiente a la función $f(x)=xsin(x)$, de linea segmentada y color cian, y la asociada la serie de Taylor de $f(x)$ centrada es $0$ hasta su décimo término, de linea continua y color magenta, se separan a partir de $x=\pi$, es decir la aproximación es buena hasta este punto.
